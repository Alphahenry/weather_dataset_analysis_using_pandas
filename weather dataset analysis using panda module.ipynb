```python
df["Date/Time"] = pd.to_datetime(df["Date/Time"])
```

```python
#setting date/time as index
df.set_index("Date/Time", inplace=True)
```

```python
#temperature plot
plt.plot(df.index,df["Temp_C"],label="temperature")
plt.xlabel("date/time")
plt.ylabel("temperature")
plt.title("Temperature")
plt.legend()
plt.show()
```

```python
#Relative humidity plot
plt.plot(df.index,df["Rel Hum_%"],label="Humidity",color="red")
plt.xlabel("date/time")
plt.ylabel("Rel Hum_%")
plt.title("weather station humidity reading")
plt.legend()
plt.show()
```

```python
#Relative humidity plot
plt.plot(df.index,df["Wind Speed_km/h"],label="windspeed",color="green")
plt.xlabel("date/time")
plt.ylabel("windspeed")
plt.title("weather station windspeed reading")
plt.legend()
plt.show()
```
